```python
# main.ipynb

# Import necessary modules and functions
import numpy as np
import pandas as pd
import pydicom
from data_loader import read_image_df, read_metadata_df, split_dataset, get_stratified_kfold
from model import create_model
from training import train_model
from evaluation import evaluate_model, cross_validate_model
from visualization import plot_confusion_matrix, plot_training_history

# Define your data directory where the labeled folders ('0' and '1') are located.
data_dir = '/Users/yavuzalpdemirci/Desktop/data_for_testing'

# 1. Create the Image DataFrame: file_path and label (extracted from folder names)
image_df = read_image_df(data_dir)
print("Image DataFrame shape:", image_df.shape)
print(image_df.head())

# 2. Create the Metadata DataFrame: file_path and metadata fields
metadata_df = read_metadata_df(data_dir)
print("Metadata DataFrame shape:", metadata_df.shape)
print(metadata_df.head())

for col in metadata_df.select_dtypes(include=['int', 'float']).columns:
    metadata_df[col] = metadata_df[col].astype('float32')

# Now, if needed, you can merge them later using the common 'file_path' key:
# merged_df = pd.merge(image_df, metadata_df, on='file_path')

# 3. Split the dataset into train (70%), validation (20%), and test (10%)
train_df, val_df, test_df = split_dataset(image_df)
print("Train/Val/Test sizes:", len(train_df), len(val_df), len(test_df))

# This function loads and processes the images with the custom metadata processing function.
def load_and_process_images(df, process_fn, metadata_df, target_shape=(512, 512)):
    """
    Loads and processes images from file paths using a custom processing function.
    Ensures all images have a consistent shape.
    
    Args:
        df (pd.DataFrame): DataFrame containing image file paths.
        process_fn (function): Function to preprocess images.
        metadata_df (pd.DataFrame): DataFrame containing metadata.
        target_shape (tuple): Desired image shape (height, width).

    Returns:
        np.ndarray: Numpy array of processed images with shape (N, 512, 512, 4).
    """
    processed_images = []
    
    for idx, row in df.iterrows():
        try:
            processed_image = process_fn(row['file_path'], metadata_df, target_shape)
            processed_images.append(processed_image)
        except Exception as e:
            print(f"Error processing {row['file_path']}: {e}")

    return np.array(processed_images) 


# Import your custom processing function (ensure process_img.py is in your PYTHONPATH)
#import sys
#sys.path.append("/Users/yavuzalpdemirci/Documents/SYZ Teknofest/project_main/")
from process_img import preprocess_with_metadata_from_df

# Load and process the images with shape checking enforced
train_images = load_and_process_images(train_df, preprocess_with_metadata_from_df, metadata_df)
train_labels = train_df['label'].values[:len(train_images)]  # Adjust label count

val_images = load_and_process_images(val_df, preprocess_with_metadata_from_df, metadata_df)
val_labels = val_df['label'].values[:len(val_images)]

test_images = load_and_process_images(test_df, preprocess_with_metadata_from_df, metadata_df)
test_labels = test_df['label'].values[:len(test_images)]

print('The dataset is defined succesfully!')

# 5. Create and summarize the model.
model = create_model(input_shape=train_images.shape[1:], num_classes=1)
model.summary()

# 6. Train the model using data augmentation on the training set.
history = train_model(model, train_images, train_labels, val_images, val_labels, batch_size=16, epochs=3)

# 7. Evaluate the model on the test set.
metrics = evaluate_model(model, test_images, test_labels)
print("Evaluation Metrics:", metrics)

# 8. Visualize the results.
plot_confusion_matrix(metrics["confusion_matrix"])
plot_training_history(history)

# (Optional) Perform cross validation using StratifiedKFold:
#skf_splits = get_stratified_kfold(image_df, n_splits=5)
#cv_scores = cross_validate_model(lambda: create_model(input_shape=train_images.shape[1:], num_classes=2), image_df, skf_splits)
#print("Cross Validation Accuracies:", cv_scores)

#print(train_labels.shape)  # Should be (num_samples,)
#print(train_labels[:10])  # Should be [0, 1, 0, 1, 1, ...]
#print(np.unique(train_labels))  # Should print [0 1]
#train_labels = np.argmax(train_labels, axis=1)
```
